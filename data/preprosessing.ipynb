```python
ratings  = pd.read_csv("/home/ananthu/machine_learning/ai_recomendation_library/data/books_data/rating.csv", on_bad_lines='warn')
ratings.head()
```

```python
ratings['User-ID'].value_counts()
```

```python
print(ratings['User-ID'].unique().shape)
```

```python
data = ratings['User-ID'].value_counts() > 200
print(data[data].shape)
```

```python
y = data[data].index
y
```

```python
ratings=ratings[ratings['User-ID'].isin(y)]
ratings.head()
```

```python
df = ratings.merge(books,on = "ISBN")
df
```

```python
num_ratings = df.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_ratings.head()
```

```python
num_ratings.rename(columns={'Book-Rating':'no_of_ratings'},inplace=True)
num_ratings.head()
```

```python
df = df.merge(num_ratings,on ='Book-Title')
df.head()
```

```python
print(df.shape)
```

```python
df = df[df['no_of_ratings']>50]
print(df.shape)
df.head()
```

```python
df.drop_duplicates(['User-ID','Book-Title'],inplace=True)
print(df.shape)
df.head()
```
