```python
import numpy as np
import pandas as pd
```

```python
df  = pd.read_csv("Book_Dataset_1.csv")
df.head()
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0','Avilability', 'Number_of_reviews','Price_After_Tax','Tax_amount'], axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
df['Category_types'] = label.fit_transform(df['Category'])
df.head()
```

```python
df['Category'].unique()
```

```python
map = dict(zip(df['Category'],df['Category_types']))
map
```

```python
sort = sorted(map.items())
sort
```

```python
data = pd.DataFrame({'category': df['Category'], 'encoded': df['Category_types']})
counts = data['encoded'].value_counts()
category_counts = counts.rename(index=lambda x: label.inverse_transform([x])[0])
print(category_counts)
```
